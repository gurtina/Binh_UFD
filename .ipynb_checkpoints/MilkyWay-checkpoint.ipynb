```python
# import modules
import numpy as np

# import plotting modules
import matplotlib
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
%matplotlib inline

from IPython.display import Latex

%matplotlib inline

from matplotlib import gridspec
from NFW_Z import NFW_Z
```

```python
bin_size = 3 # in kpc
starting_radius = 92170 # in kpc
```

```python
MilkyWay = np.loadtxt("D:/Research/ultra-faint dwarf galaxies/Milky Way/742935_dm_properties_TNGv2.txt")
```

```python
dx_MilkyWay = MilkyWay[:,0] - 33960.34 # in kpc
dy_MilkyWay = MilkyWay[:,1] - 70655.805 # in kpc
dz_MilkyWay = MilkyWay[:,2] - 2442.741 # in kpc

dx_MilkyWay_shifted = [x - 33960.34 for x in dx_MilkyWay] # shifted to the halo's center of mass
dy_MilkyWay_shifted = [y - 70655.805 for y in dy_MilkyWay] # shifted to the halo's center of mass
dz_MilkyWay_shifted = [x - 2442.741 for z in dz_MilkyWay] # shifted to the halo's center of mass

r_MilkyWay = np.sqrt(dx_MilkyWay_shifted**2 + dy_MilkyWay_shifted**2 + dz_MilkyWay_shifted**2)
# max: 92183.53271005189 kpc, closest to 92190 kpc
print(max(r_MilkyWay))
print(r_MilkyWay)
rad_array_MilkyWay = np.arange(starting_radius, 92190, bin_size) # unit: kpc
```

```python
density_profile_MilkyWay = np.zeros(np.size(rad_array_MilkyWay))
i = 0
for radius_value in rad_array_MilkyWay:
    particles = np.where((r_MilkyWay > radius_value) & (r_MilkyWay < (radius_value + bin_size)))
    how_many_particles = np.size(particles)
    print(how_many_particles, 'how_many_particles')
    print(radius_value + bin_size, 'radius_value + bin_size')
    print(radius_value, 'radius_value')
    shell_volume = 4/3*np.pi*((radius_value + bin_size)**3 - radius_value**3)
    print((radius_value + bin_size)**3 - radius_value**3, '(radius_value + bin_size)**3 - radius_value**3')
    print(shell_volume, 'shell_volume')
    density_profile_MilkyWay[i] = how_many_particles / shell_volume * 6.3e6
    print(density_profile_MilkyWay[i], 'density profile\n')
    i += 1
```

```python
mass_profile_MilkyWay = np.zeros(np.size(rad_array_MilkyWay))
h = 0
for radius_value in rad_array_MilkyWay:
    particles = np.where((r_MilkyWay < (radius_value + bin_size)))
    how_many_particles = np.size(particles)
    mass_profile_MilkyWay[h] = how_many_particles * 6.3e6
    h += 1
```

```python
fig, ax = plt.subplots(figsize=[10, 10])
plt.plot(rad_array_MilkyWay, density_profile_MilkyWay, color=(0.5, 0.1, 0.9, 1), linewidth=2, \
         label="average density profile")
plt.xlabel('distance from galactic center (kpc)', fontsize=17)
plt.ylabel('average dark matter density ($M_\odot$/kpc$^3$)', fontsize=17) # using LATEX here!
plt.title('average dark matter density profile of the Milky Way', loc='center', fontsize=19)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.show()
#plt.xscale('log')
#plt.yscale('log')   
```
