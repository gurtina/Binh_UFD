```python
import numpy as np

import matplotlib.patches as mpatches
import matplotlib.pyplot as plt

from IPython.display import Latex

%matplotlib inline

from NFW import NFW
```

```python
# this works for halo 1

halo1 = NFW(1e9)
r_s1 = halo1.r_s()
c_vir1 = halo1.c_vir()
print('r_s =', r_s1)
print('c_vir =', c_vir1)

radius1 = np.arange(1, 21, 1) #unit: kpc
density_profile1 = halo1.rho(radius1)
```

```python
# this works for halo 2

halo2 = NFW(1e10)
r_s2 = halo2.r_s()
c_vir2 = halo2.c_vir()
print('r_s =', r_s2)
print('c_vir =', c_vir2)

radius2 = np.arange(1, 21, 1) #unit: kpc
density_profile2 = halo2.rho(radius2)
```

```python
# this works for halo 3

halo3 = NFW(1e11)
r_s3 = halo3.r_s()
c_vir3 = halo3.c_vir()
print('r_s =', r_s3)
print('c_vir =', c_vir3)

radius3 = np.arange(1, 21, 1) #unit: kpc
density_profile3 = halo3.rho(radius3)
```

```python
# this works for halo 4

halo4 = NFW(1e12)
r_s4 = halo4.r_s()
c_vir4 = halo4.c_vir()
print('r_s =', r_s4)
print('c_vir =', c_vir4)

radius4 = np.arange(1, 21, 1) #unit: kpc
density_profile4 = halo4.rho(radius4)
```

```python
# this works for halo 5, virial mass = 1.903021e+08 solar masses. This is dwarf 1 at z = 0.130

halo5 = NFW(1.903021e+08)
r_s5 = halo5.r_s()
c_vir5 = halo5.c_vir()
print('r_s =', r_s5)
print('c_vir =', c_vir5)

radius5 = np.arange(1, 21, 1) #unit: kpc
density_profile5 = halo5.rho(radius5)
```

```python
plt.figure(figsize=(8,8))

plt.plot(radius1, density_profile1, color='#8d6a9f', linewidth=2)
plt.plot(radius2, density_profile2, color='#c5cbd3', linewidth=2)
plt.plot(radius3, density_profile3, color='#dda448', linewidth=2)
plt.plot(radius4, density_profile4, color='#bb342f', linewidth=2)
plt.plot(radius5, density_profile5, color='black', linewidth=2)

line1 = mpatches.Patch(color='#8d6a9f', label=r'virial mass = 10e9 M$_O$')
line2 = mpatches.Patch(color='#c5cbd3', label=r'virial mass = 10e10 M$_O$')
line3 = mpatches.Patch(color='#dda448', label=r'virial mass = 10e11 M$_O$')
line4 = mpatches.Patch(color='#bb342f', label=r'virial mass = 10e12 M$_O$')
line5 = mpatches.Patch(color='black', label=r'virial mass = 1.903021e+08 M$_O$')

legend = plt.legend(handles=[line1, line2, line3, line4, line5],loc='lower left')

plt.xlabel('radius (kpc)', fontsize=15)
plt.ylabel(r'dark matter density profile (particles/kpc$^3$)', fontsize=15)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.xscale('log')
plt.yscale('log')

plt.show()
```
