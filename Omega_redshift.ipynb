```python
import numpy as np

import matplotlib.patches as mpatches
import matplotlib.pyplot as plt

from IPython.display import Latex

%matplotlib inline

from NFW_Z import NFW_Z
```

```python
halo1 = NFW_Z(1e9, "Jeon")

redshift = np.arange(1, 101, 1)
OmegaM_Z = halo1.OmegaM_Z(redshift)
OmegaR_Z = halo1.OmegaR_Z(redshift)
OmegaL_Z = halo1.OmegaL_Z(redshift)
```

```python
plt.figure(figsize=(10,10))

plt.plot(redshift, OmegaM_Z, color='red', linewidth=2, label='Ω$_M$')
plt.plot(redshift, OmegaR_Z, color='green', linewidth=2, label='Ω$_R$')
plt.plot(redshift, OmegaL_Z, color='blue', linewidth=2, label='Ω$_L$')

plt.legend(bbox_to_anchor=(1.02, 1.01), loc='upper left')

plt.xlabel('z', fontsize=17)
plt.ylabel('Ω', fontsize=17) # using LATEX here!
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.title('change in Ω over multiple redshifts', loc='center', fontsize=20)

plt.show()
```
