```python
import numpy as np
import matplotlib.pyplot as plt

data = np.genfromtxt('D:/Research/ultra-faint dwarf galaxies/observational_data/r05_ufd.txt',dtype='str')
r05_ufd = np.array(data[:,2], dtype='float32')
sm_ufd = np.array(data[:,3], dtype='float32')

r05_ufd_sim1_noRT = np.array([0.19, 0.17, 0.17]) * 1000 #kpc into pc
r05_ufd_sim1_RT = np.array([0.15, 0.15, 0.21]) * 1000 #kpc into pc
sm_ufd_sim1_noRT = np.array([8769.38, 8778.7385, 8778.983]) #M_sun
sm_ufd_sim1_RT = np.array([3247.37, 3247.452, 3247.549]) #M_sun
r05_ufd_sim2_noRT = np.array([0.39, 0.35, 0.45]) * 1000 #kpc into pc
r05_ufd_sim2_RT = np.array([0.03, 0.03, 0.05]) * 1000 #kpc into pc
sm_ufd_sim2_noRT = np.array([3051.31, 3051.3835, 3051.5635]) * 2 #M_sun
sm_ufd_sim2_RT = np.array([1101.5865, 1101.6125000000002, 1101.6795]) * 2 #M_sun

plt.scatter(np.log10(sm_ufd*1e6), np.log10(r05_ufd), marker='^', s=130, edgecolor='black', 
            color='blue', alpha=0.6, label='MW UFDs')
plt.scatter(np.log10(sm_ufd_sim1_noRT), np.log10(r05_ufd_sim1_noRT), marker='o', s=130, edgecolor='black', 
            color='pink', alpha=0.6, label='UFD1, no RT')
plt.scatter(np.log10(sm_ufd_sim1_RT), np.log10(r05_ufd_sim1_RT), marker='o', s=130, edgecolor='black', 
            color='red', alpha=0.6, label='UFD1, RT')
plt.scatter(np.log10(sm_ufd_sim2_noRT), np.log10(r05_ufd_sim2_noRT), marker='s', s=130, edgecolor='black', 
            color='pink', alpha=0.6, label='UFD2, no RT')
plt.scatter(np.log10(sm_ufd_sim2_RT), np.log10(r05_ufd_sim2_RT), marker='s', s=130, edgecolor='black', 
            color='red', alpha=0.6, label='UFD2, RT')

data = np.genfromtxt('D:/Research/ultra-faint dwarf galaxies/observational_data/r05_dsph.txt',dtype='str')
r05_dsph = np.array(data[:,2], dtype='float32')
sm_dsph = np.array(data[:,3], dtype='float32')
plt.scatter(np.log10(sm_dsph*1e6), np.log10(r05_dsph), marker='^', s=130, edgecolor='black', 
            color='black', alpha=0.6, label='MW dSphs')

plt.legend(loc='lower right')

plt.xlabel('log M$_{*}$ (M$_\odot$)')
plt.ylabel('log r$_{1/2}$ [pc]')
```
