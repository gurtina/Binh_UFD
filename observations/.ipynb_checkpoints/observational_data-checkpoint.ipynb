```python
import numpy as np
import matplotlib.pyplot as plt

data = np.genfromtxt('D:/Research/ultra-faint dwarf galaxies/observational_data/r05_ufd.txt',dtype='str')
r05_ufd = np.array(data[:,2], dtype='float32')
sm_ufd = np.array(data[:,3], dtype='float32')

r05_ufd_sim1 = np.array([0.19, 0.15, 0.17, 0.15, 0.17, 0.21]) * 1000 #kpc into pc
sm_ufd_sim1 = np.array([8769.38, 3247.37, 8778.7385, 3247.452, 8778.983, 3247.549]) #M_sun
r05_ufd_sim2 = np.array([0.39, 0.03, 0.35, 0.03, 0.45, 0.05]) * 1000 #kpc into pc
sm_ufd_sim2 = np.array([3051.31, 1101.5865, 3051.3835, 1101.6125000000002, \
                        3051.5635, 1101.6795]) #M_sun

plt.scatter(np.log10(sm_ufd*1e6), np.log10(r05_ufd), marker='^', s=130, edgecolor='black', 
            color='blue', alpha=0.6, label='MW UFDs')
plt.scatter(np.log10(sm_ufd_sim1*1e6), np.log10(r05_ufd_sim1), marker='o', s=130, edgecolor='black', 
            color='blue', alpha=0.6, label='simulated UFD #1')
plt.scatter(np.log10(sm_ufd_sim2*1e6), np.log10(r05_ufd_sim2), marker='P', s=130, edgecolor='black', 
            color='blue', alpha=0.6, label='simulated UFD #2')

data = np.genfromtxt('D:/Research/ultra-faint dwarf galaxies/observational_data/r05_dsph.txt',dtype='str')
r05_dsph = np.array(data[:,2], dtype='float32')
sm_dsph = np.array(data[:,3], dtype='float32')
plt.scatter(np.log10(sm_dsph*1e6), np.log10(r05_dsph), marker='^', s=130, edgecolor='black', 
            color='black', alpha=0.6, label='MW dSphs')

plt.legend(loc='lower center')

plt.xlabel('sm')
plt.ylabel('r05')
```
