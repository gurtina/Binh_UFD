```python
# import modules
import numpy as np

# import plotting modules
import matplotlib
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
%matplotlib inline

from IPython.display import Latex

%matplotlib inline
```

```python
analyze_300 = np.loadtxt("D:/Research/ultra-faint dwarf galaxies/velocity/dwarf1/RT/dwarf_dm_5e8_rt_0300.txt")

dx300 = analyze_300[:,4] / 1000 # converting pc to kpc
dy300 = analyze_300[:,5] / 1000 # converting pc to kpc
dz300 = analyze_300[:,6] / 1000 # converting pc to kpc
r300 = np.sqrt(dx300**2 + dy300**2 + dz300**2)

vx300 = analyze_300[:,7]
vy300 = analyze_300[:,8]
vz300 = analyze_300[:,9]
vx_COM = np.sum(vx300) / np.size(vx300)
vy_COM = np.sum(vy300) / np.size(vy300)
vz_COM = np.sum(vz300) / np.size(vz300)
vx_new = vx300 - vx_COM
vy_new = vy300 - vy_COM
vz_new = vz300 - vz_COM
v_new = np.sqrt(vx_new**2 + vy_new**2 + vz_new**2)
```

```python
fig, ax = plt.subplots(figsize=[10, 10])

plt.plot(r300, v_new, '.', color='red')

plt.xlabel('distance from galactic center (kpc)', fontsize=17)
plt.ylabel('velocity (km/s)', fontsize=17) # using LATEX here!
plt.title('speed of particle with respect to distance\nfrom dwarf center, dwarf 1, z = 0.024, with RT', loc='center', fontsize=17)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.show()
```
