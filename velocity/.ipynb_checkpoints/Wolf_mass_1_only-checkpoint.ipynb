```python
fig, ax = plt.subplots(figsize=[10, 10])

halo1noRT130, = ax.plot('no RT,\nz = 0.130', M_half_DM[0] / M_half_star[0], 's', color='#3185FC', markersize=11)
plt.plot('no RT,\nz = 0.130', M_half_x_wolf[0] / M_half_star[0], 's', color='#083D77', markersize=11)
plt.plot('no RT,\nz = 0.130', M_half_y_wolf[0] / M_half_star[0], 's', color='#DA4167', markersize=11)
plt.plot('no RT,\nz = 0.130', M_half_z_wolf[0] / M_half_star[0], 's', color='#F4D35E', markersize=11)

halo1RT130, = ax.plot('with RT,\nz = 0.130', M_half_DM[1] / M_half_star[1], '^', color='#3185FC', markersize=11)
plt.plot('with RT,\nz = 0.130', M_half_x_wolf[1] / M_half_star[1], '^', color='#083D77', markersize=11)
plt.plot('with RT,\nz = 0.130', M_half_y_wolf[1] / M_half_star[1], '^', color='#DA4167', markersize=11)
plt.plot('with RT,\nz = 0.130', M_half_z_wolf[1] / M_half_star[1], '^', color='#F4D35E', markersize=11)

halo1noRT245, = ax.plot('no RT,\nz = 0.245', M_half_DM[2] / M_half_star[2], 'o', color='#3185FC', markersize=11)
plt.plot('no RT,\nz = 0.245', M_half_x_wolf[2] / M_half_star[2], 'o', color='#083D77', markersize=11)
plt.plot('no RT,\nz = 0.245', M_half_y_wolf[2] / M_half_star[2], 'o', color='#DA4167', markersize=11)
plt.plot('no RT,\nz = 0.245', M_half_z_wolf[2] / M_half_star[2], 'o', color='#F4D35E', markersize=11)

halo1RT245, = ax.plot('with RT,\nz = 0.245', M_half_DM[3] / M_half_star[3], 'P', color='#3185FC', markersize=11)
plt.plot('with RT,\nz = 0.245', M_half_x_wolf[3] / M_half_star[3], 'P', color='#083D77', markersize=11)
plt.plot('with RT,\nz = 0.245', M_half_y_wolf[3] / M_half_star[3], 'P', color='#DA4167', markersize=11)
plt.plot('with RT,\nz = 0.245', M_half_z_wolf[3] / M_half_star[3], 'P', color='#F4D35E', markersize=11)

halo1noRT373, = ax.plot('no RT,\nz = 0.373', M_half_DM[4] / M_half_star[4], 'd', color='#3185FC', markersize=11)
plt.plot('no RT,\nz = 0.373', M_half_x_wolf[4] / M_half_star[4], 'd', color='#083D77', markersize=11)
plt.plot('no RT,\nz = 0.373', M_half_y_wolf[4] / M_half_star[4], 'd', color='#DA4167', markersize=11)
plt.plot('no RT,\nz = 0.373', M_half_z_wolf[4] / M_half_star[4], 'd', color='#F4D35E', markersize=11)

halo1RT373, = ax.plot('with RT,\nz = 0.373', M_half_DM[5] / M_half_star[5], 'h', color='#3185FC', markersize=11)
plt.plot('with RT,\nz = 0.373', M_half_x_wolf[5] / M_half_star[5], 'h', color='#083D77', markersize=11)
plt.plot('with RT,\nz = 0.373', M_half_y_wolf[5] / M_half_star[5], 'h', color='#DA4167', markersize=11)
plt.plot('with RT,\nz = 0.373', M_half_z_wolf[5] / M_half_star[5], 'h', color='#F4D35E', markersize=11)

#plt.xlabel('redshift', fontsize=17)
plt.ylabel(r'$\frac{M_{half,DM}}{M_{half,stars}}$', fontsize=24) # using LATEX here!

plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
patch1 = mpatches.Patch(color='#3185FC', label='DM halo')
patch2 = mpatches.Patch(color='#083D77', label='Wolf, x')
patch3 = mpatches.Patch(color='#DA4167', label='Wolf, y')
patch4 = mpatches.Patch(color='#F4D35E', label='Wolf, z')

#first_legend = ax.legend(handles=[halo1noRT130, halo1noRT245, halo1noRT373,\
#                                  halo1RT130, halo1RT245, halo1RT373], fontsize=16, loc='upper center')

#ax = plt.gca().add_artist(first_legend)

plt.legend(handles=[patch1, patch2, patch3, patch4], fontsize=16, loc='upper center')

plt.title('Ratio between M$_{half,DM}$ and M$_{half,stars}$\nwith respect to M$_{half,stars}$', \
         fontsize=19)

plt.yscale('log')

plt.show()
```
