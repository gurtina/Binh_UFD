```python
# import modules
import numpy as np

# import plotting modules
import matplotlib
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
```

```python
r_half = np.array([0.17, 0.13, 0.49, 0.05])
M_half_star = np.array([8779.301, 3247.6735, 3051.688, 1101.59])
M_half_DM = np.array([5116500, 2939000, 14014000, 357000])
M_half_x_wolf = np.array([1968934.751, 1165376.999, 847206.2754, 79233.88201])
M_half_y_wolf = np.array([878251.727, 896219.1646, 3858987.116, 11695.22367])
M_half_z_wolf = np.array([1939348.312, 4373286.711, 2054710.263, 134120.7022])
```

```python
fig, ax = plt.subplots(figsize=[10, 10])

halo1noRT, = ax.plot(M_half_star[0], M_half_DM[0] / M_half_star[0], 's', color='#3185FC', markersize=11, \
                     label='dwarf 1 no RT, z = 0.514')
plt.plot(M_half_star[0], M_half_x_wolf[0] / M_half_star[0], 's', color='#083D77', markersize=11)
plt.plot(M_half_star[0], M_half_y_wolf[0] / M_half_star[0], 's', color='#DA4167', markersize=11)
plt.plot(M_half_star[0], M_half_z_wolf[0] / M_half_star[0], 's', color='#F4D35E', markersize=11)

halo1RT, = ax.plot(M_half_star[1], M_half_DM[1] / M_half_star[1], '^', color='#3185FC', markersize=11, \
                   label='dwarf 1 RT, z = 0.514')
plt.plot(M_half_star[1], M_half_x_wolf[1] / M_half_star[1], '^', color='#083D77', markersize=11)
plt.plot(M_half_star[1], M_half_y_wolf[1] / M_half_star[1], '^', color='#DA4167', markersize=11)
plt.plot(M_half_star[1], M_half_z_wolf[1] / M_half_star[1], '^', color='#F4D35E', markersize=11)

halo2noRT, = ax.plot(M_half_star[2], M_half_DM[2] / M_half_star[2], 'o', color='#3185FC', markersize=11, \
                     label='dwarf 2 no RT, z = 0.514')
plt.plot(M_half_star[2], M_half_x_wolf[2] / M_half_star[2], 'o', color='#083D77', markersize=11)
plt.plot(M_half_star[2], M_half_y_wolf[2] / M_half_star[2], 'o', color='#DA4167', markersize=11)
plt.plot(M_half_star[2], M_half_z_wolf[2] / M_half_star[2], 'o', color='#F4D35E', markersize=11)

halo2RT, = ax.plot(M_half_star[3], M_half_DM[3] / M_half_star[3], 'P', color='#3185FC', markersize=11, \
                   label='dwarf 2 RT, z = 0.514')
plt.plot(M_half_star[3], M_half_x_wolf[3] / M_half_star[3], 'P', color='#083D77', markersize=11)
plt.plot(M_half_star[3], M_half_y_wolf[3] / M_half_star[3], 'P', color='#DA4167', markersize=11)
plt.plot(M_half_star[3], M_half_z_wolf[3] / M_half_star[3], 'P', color='#F4D35E', markersize=11)

plt.xlabel('M$_{half,stars}$ (M$_\odot$)', fontsize=17)
plt.ylabel(r'$\frac{M_{half,DM}}{M_{half,stars}}$', fontsize=24) # using LATEX here!

plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
patch1 = mpatches.Patch(color='#3185FC', label='DM halo')
patch2 = mpatches.Patch(color='#083D77', label='Wolf, x')
patch3 = mpatches.Patch(color='#DA4167', label='Wolf, y')
patch4 = mpatches.Patch(color='#F4D35E', label='Wolf, z')

first_legend = ax.legend(handles=[halo1noRT, halo1RT, halo2noRT, halo2RT], fontsize=18, loc='upper right')

ax = plt.gca().add_artist(first_legend)

plt.legend(handles=[patch1, patch2, patch3, patch4], fontsize=18, loc='lower right')

plt.yscale('log')

plt.show()
```
