```python
# import modules
import numpy as np

# import plotting modules
import matplotlib
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
%matplotlib inline

from IPython.display import Latex

%matplotlib inline

from NFW import NFW
```

```python
dr = 10
```

```python
'''THIS PART WORKS WITH NO_RT'''

전data290_noRT = np.loadtxt("D:/Research/ultra-faint dwarf galaxies/dwarf 1 no RT/dwarf_dm_5e8_0290.txt")
dx290_noRT = 전data290_noRT[:,4]
dy290_noRT = 전data290_noRT[:,5]
dz290_noRT = 전data290_noRT[:,6]
r290_noRT = np.sqrt(dx290_noRT**2 + dy290_noRT**2 + dz290_noRT**2) # max: 11333.935323605565, closest to 11340
rad_array290_noRT = np.arange(10, 11340, dr, dtype=np.int64)
density_profile290_noRT = np.zeros(np.size(rad_array290_noRT))
i = 0
for radius_value in rad_array290_noRT:
    particles = np.where((r290_noRT > radius_value) & (r290_noRT < (radius_value + dr)))
    how_many_particles = np.size(particles)
    shell_volume = 4/3*np.pi*((radius_value + dr)**3 - radius_value**3)
    density_profile290_noRT[i] = how_many_particles / shell_volume
    i += 1
densityprof290_noRT = density_profile290_noRT / density_profile290_noRT[0]
```

```python
# pick a halo mass, change the concentration
# dwarf 1, redshift z = 0.130, halo mass = 1.903021e+08 solar masses

halo1 = NFW(1.903021e+08)
r_s1 = halo1.r_s()
c_vir1 = halo1.c_vir()
print('r_s =', r_s1)
print('c_vir =', c_vir1)

radius1 = np.arange(0.01, 11.34, 0.01) #unit: kpc
density_profile1 = halo1.rho(radius1, c=11)
NFWdensityprof_noRT = density_profile1 / density_profile1[0]
print(NFWdensityprof_noRT)
```

```python
plt.figure(figsize=(8,8))

plt.plot(rad_array290_noRT, densityprof290_noRT, color=(0.5, 0.1, 0.9, 1), linewidth=2)
plt.plot(rad_array290_noRT, NFWdensityprof_noRT, color='#8d6a9f', linewidth=2)

plt.xlabel('distance from center of galaxy halo (pc)', fontsize=17)
plt.ylabel('normalized dark matter density', fontsize=17) # using LATEX here!
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
#plt.xlim(10,10**4.2)
plt.xscale('log')
plt.yscale('log')

plt.title('normalized dark matter density profile of dwarf 1, high z, without RT,\nbin size = ' \
          + str(dr) + ', plotted with corresponding NFW profile', loc='center', fontsize=17)

plt.show()
```

```python
'''THIS PART WORKS WITH RT'''

전data290 = np.loadtxt("D:/Research/ultra-faint dwarf galaxies/dwarf 1 RT/dwarf_dm_5e8_rt_0290.txt")
dx290 = 전data290[:,4]
dy290 = 전data290[:,5]
dz290 = 전data290[:,6]
r290 = np.sqrt(dx290**2 + dy290**2 + dz290**2) # max: 11415.825965662318, closest to 11420
rad_array290 = np.arange(10, 11420, dr, dtype=np.int64)
density_profile290 = np.zeros(np.size(rad_array290))
i = 0
for radius_value in rad_array290:
    particles = np.where((r290 > radius_value) & (r290 < (radius_value + dr)))
    how_many_particles = np.size(particles)
    shell_volume = 4/3*np.pi*((radius_value + dr)**3 - radius_value**3)
    density_profile290[i] = how_many_particles / shell_volume
    i += 1
densityprof290 = density_profile290 / density_profile290[0]
```

```python
plt.figure(figsize=(13,13))

plt.plot(rad_array290, densityprof290, color=(0.2, 0.9, 0.6, 1), linewidth=2)

plt.xlabel('distance from center of galaxy halo (pc)', fontsize=17)
plt.ylabel('normalized dark matter density', fontsize=17) # using LATEX here!
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlim(10,10**4.2)
plt.xscale('log')
plt.yscale('log')

plt.title('normalized dark matter density profile of dwarf 1, high z, with RT,\nbin size = ' \
          + str(dr) + ', plotted with corresponding NFW profile', loc='center', fontsize=17)

plt.show()
```
