```python
# now, we calculate the difference between the density profiles at each redshift with and without RT.
```

```python
# compare the size of the arrays with and without RT

arraylist = [densityprof290_noRT, densityprof290, \
             densityprof270_noRT, densityprof270, \
             densityprof250_noRT, densityprof250, \
             densityprof230_noRT, densityprof230, \
             densityprof210_noRT, densityprof210, \
             densityprof190_noRT, densityprof190, \
             densityprof180_noRT, densityprof180]
i = 0
while i < len(arraylist):
    difference = np.size(arraylist[i]) - np.size(arraylist[i+1])
    if difference < 0:
        arraylist[i] = np.concatenate([arraylist[i], np.zeros(abs(difference))])
        print('with RT longer')
        i += 2
    else:
        arraylist[i+1] = np.concatenate([arraylist[i+1], np.zeros(abs(difference))])
        print('without RT longer')
        i += 2
        
diff290 = arraylist[0] - arraylist[1]
diff270 = arraylist[2] - arraylist[3]
diff250 = arraylist[4] - arraylist[5]
diff230 = arraylist[6] - arraylist[7]
diff210 = arraylist[8] - arraylist[9]
diff190 = arraylist[10] - arraylist[11]
diff180 = arraylist[12] - arraylist[13]
```
