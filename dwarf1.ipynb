```python
전data290 = np.loadtxt("D:/Research/ultra-faint dwarf galaxies/dwarf 1 RT/dwarf_dm_5e8_rt_0290.txt")
dx290 = 전data290[:,4]
dy290 = 전data290[:,5]
dz290 = 전data290[:,6]
r290 = np.sqrt(dx290**2 + dy290**2 + dz290**2) # max: 11415.825965662318, closest to 11420
rad_array290 = np.arange(10, 11420, dr, dtype=np.int64)
density_profile290 = np.zeros(np.size(rad_array290))
i = 0
for radius_value in rad_array290:
    particles = np.where((r290 > radius_value) & (r290 < (radius_value + dr)))
    how_many_particles = np.size(particles)
    shell_volume = 4/3*np.pi*((radius_value + dr)**3 - radius_value**3)
    density_profile290[i] = how_many_particles / shell_volume
    i += 1
```

```python
# compare the size of the arrays with and without RT

arraylist = [density_profile290_noRT, density_profile290, density_profile280_noRT, density_profile280, \
             density_profile270_noRT, density_profile270, density_profile260_noRT, density_profile260, \
             density_profile250_noRT, density_profile250, density_profile240_noRT, density_profile240, \
             density_profile230_noRT, density_profile230, density_profile220_noRT, density_profile220, \
             density_profile210_noRT, density_profile210, density_profile200_noRT, density_profile200, \
             density_profile190_noRT, density_profile190, density_profile180_noRT, density_profile180, \
             density_profile170_noRT, density_profile170, density_profile160_noRT, density_profile160, \
             density_profile150_noRT, density_profile150, density_profile140_noRT, density_profile140, \
             density_profile130_noRT, density_profile130, density_profile120_noRT, density_profile120, \
             density_profile110_noRT, density_profile110, density_profile100_noRT, density_profile100, \
             density_profile99_noRT, density_profile99, density_profile98_noRT, density_profile98, \
             density_profile97_noRT, density_profile97, density_profile96_noRT, density_profile96, \
             density_profile95_noRT, density_profile95, density_profile94_noRT, density_profile94, \
             density_profile93_noRT, density_profile93, density_profile92_noRT, density_profile92, \
             density_profile91_noRT, density_profile91, density_profile71_noRT, density_profile71, \
             density_profile51_noRT, density_profile51]

i = 0
while i < len(arraylist):
    difference = np.size(arraylist[i]) - np.size(arraylist[i+1])
    if difference < 0:
        arraylist[i] = np.concatenate([arraylist[i], np.zeros(abs(difference))])
        print('RT expands further')
        i += 2
    elif difference > 0:
        arraylist[i+1] = np.concatenate([arraylist[i+1], np.zeros(abs(difference))])
        print('no_RT expands further')
        i += 2
    else:
        print('expands to the same size')
        i += 2

# diff = no_RT - RT for each redshift
        
j = 0
while i < len(arraylist):
    diff = arraylist[j] - arraylist[j+1]
    diff_list.append(diff)
    j += 2
```
