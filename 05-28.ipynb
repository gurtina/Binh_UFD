```python
import numpy as np

import matplotlib.pyplot as plt

from IPython.display import Latex

%matplotlib inline

from NFW import NFW
```

```python
halo = NFW(1e9)
```

```python
halo.Mvir/1e9
```

```python
halo.v_vir()
```

```python
halo.r_vir()
```

```python
halo.r_s()
```

```python
halo.c_vir()
```

```python
x=np.arange(1, 21, 1) #unit: kpc
```

```python
t=halo.rho(x)
```

```python
plt.plot(x,t)
plt.xlabel('r (kpc)')
plt.ylabel('rho(r)')
plt.xscale('log')
plt.yscale('log')
plt.show()
```

```python
y=halo.mass(x)
```

```python
plt.plot(x,y)
plt.xlabel('r (kpc)')
plt.ylabel('mass(r)')
plt.yscale('log')
plt.show()
```
