```python
import numpy as np

import matplotlib.pyplot as plt

from IPython.display import Latex

%matplotlib inline

from NFW import NFW
```

```python
halo = NFW(1e9)
```

```python
halo.Mvir/1e9
```

```python
halo.v_vir()
```

```python
halo.r_vir()
```

```python
halo.r_s()
```

```python
halo.c_vir()
```

```python
x=np.arange(1, 21, 1) #unit: kpc
```

```python
t=halo.rho(x)
```

```python
plt.plot(x,t)
plt.xlabel('r (kpc)')
plt.ylabel('rho(r)')
plt.xscale('log')
plt.yscale('log')
plt.show()
```

```python
y=halo.mass(x)
```

```python
plt.plot(x,y)
plt.xlabel('r (kpc)')
plt.ylabel('mass(r)')
plt.yscale('log')
plt.show()
```

```python
#This is to overplot the power laws with r^-1 and r^-3.

#plt.figure(figsize=(10,10))

plt.plot(x,t)
plt.xlabel('r (kpc)')
plt.ylabel('rho(r)')
C = 6.38 #remember C=7.109
plt.plot(x,10**C/x)
plt.plot(x,10**C/x**3)
plt.axvline(x=1.5795373383751166)
plt.xscale('log')
plt.yscale('log')
#plt.xlim(0.5,5)
#plt.ylim(10**7,10**7.2)
plt.show()
```

```python
plt.plot(x,t,lw=3)
plt.xlabel('r (kpc)')
plt.ylabel('rho(r)')
for C in np.arange(6.5, 6.9, 0.1):
    plt.plot(x,10**C/x,lw=0.8)
    plt.plot(x,10**C/x**3,lw=0.8)
plt.axvline(x=1.5795373383751166)
plt.xscale('log')
plt.yscale('log')
plt.show()
```
