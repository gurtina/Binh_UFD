```python
import numpy as np

import matplotlib.pyplot as plt

from IPython.display import Latex

%matplotlib inline

from NFW import NFW
```

```python
halo = NFW(1e9)
```

```python
halo.Mvir/1e9
```

```python
halo.v_vir()
```

```python
halo.r_vir()
```

```python
halo.r_s()
```

```python
halo.c_vir()
```

```python
x=np.arange(1, 21, 1) #unit: kpc
```

```python
d=halo.rho(x) #this part is to plot the density profile
plt.plot(x,d)
plt.xlabel('r (kpc)')
plt.yscale('log')
plt.ylabel('rho(r)')
plt.show()
```

```python
m=halo.mass(x) #this part is to plot the mass profile
plt.plot(x,m)
plt.xlabel('r (kpc)')
plt.yscale('log')
plt.ylabel('mass(r)')
plt.show()
```

```python
#this part is to plot the density profile again but as the mass profile divided by the volume at each radius,
#only to make sure that the rho(x) function is correct.

q=halo.mass(x)/(4/3*np.pi*x**3)
plt.plot(x,q)
plt.xlabel('r (kpc)')
plt.yscale('log')
plt.ylabel('density(r)')
plt.show()
```
