```python
import numpy as np

import matplotlib.patches as mpatches
import matplotlib.pyplot as plt

from IPython.display import Latex

%matplotlib inline

from NFW import NFW
```

```python
# this works for halo 1

halo1 = NFW(1e9)
r_s1 = halo1.r_s()
c_vir1 = halo1.c_vir()
print('r_s =', r_s1)
print('c_vir =', c_vir1)

radius1 = np.arange(0.1, 100, 0.5) #unit: kpc
density_profile1 = halo1.rho(radius1)
normalized_dens1 = density_profile1 / density_profile1[0]
```

```python
# this works for halo 2

halo2 = NFW(1e10)
r_s2 = halo2.r_s()
c_vir2 = halo2.c_vir()
print('r_s =', r_s2)
print('c_vir =', c_vir2)

radius2 = np.arange(0.1, 100, 0.5) #unit: kpc
density_profile2 = halo2.rho(radius2)
normalized_dens2 = density_profile2 / density_profile2[0]
```

```python
# this works for halo 5, virial mass = 1.903021e+08 solar masses. This is dwarf 1 at z = 0.130

halo5 = NFW(1.903021e+08)
r_s5 = halo5.r_s()
c_vir5 = halo5.c_vir()
print('r_s =', r_s5)
print('c_vir =', c_vir5)

radius5 = np.arange(0.1, 100, 0.5) #unit: kpc
density_profile5 = halo5.rho(radius5)
normalized_dens5 = density_profile5 / density_profile5[0]
```

```python
plt.figure(figsize=(8,8))

plt.plot(radius1, normalized_dens1, color='#8d6a9f', linewidth=2)
plt.plot(radius2, normalized_dens2, color='#c5cbd3', linewidth=2)
plt.plot(radius3, normalized_dens3, color='#dda448', linewidth=2)
plt.plot(radius4, normalized_dens4, color='#bb342f', linewidth=2)
plt.plot(radius5, normalized_dens5, color='black', linewidth=2)

line1 = mpatches.Patch(color='#8d6a9f', label=r'M_vir = 10e9 M$_\odot$')
line2 = mpatches.Patch(color='#c5cbd3', label=r'M_vir = 10e10 M$_\odot$')
line3 = mpatches.Patch(color='#dda448', label=r'M_vir = 10e11 M$_\odot$')
line4 = mpatches.Patch(color='#bb342f', label=r'M_vir = 10e12 M$_\odot$')
line5 = mpatches.Patch(color='black', label=r'M_vir = 1.903021e+08 M$_\odot$')

legend = plt.legend(handles=[line1, line2, line3, line4, line5],loc='lower left')

plt.xlabel('radius (kpc)', fontsize=15)
plt.ylabel(r'dark matter density profile (particles/kpc$^3$)', fontsize=15)
plt.xlim(0.1,100)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.xscale('log')
plt.yscale('log')

plt.show()
```
