```python
# import modules
import numpy as np

# import plotting modules
import matplotlib
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
```

```python
r_half = np.array([0.39, 0.03, 0.35, 0.03, 0.45, 0.05])
M_half_star = np.array([3051.31, 1101.5865, 3051.3835, 1101.6125000000002, \
                        3051.5635, 1101.6795])
M_half_DM = np.array([10123000.0, 3705500.0, 12520500.0, 2135500.0, \
                     10856500.0, 391500.0])
M_half_x_wolf = np.array([1985054.483010112, 147670.92549669172, 4096783.8230332304, 910965.2028056713, \
                         716224.9263799542, 42021.82907673496])
M_half_y_wolf = np.array([3335240.8153377315, 278136.44404964685, 2214361.1066262783, 196983.11093256588, \
                         788965.1924907799, 98123.47927289609])
M_half_z_wolf = np.array([2223017.4982447335, 252506.2723882034, 2574720.047255256, 721689.001806175, \
                         1743456.9475306368, 91791.93837654445])
```

```python
fig, ax = plt.subplots(figsize=[10, 10])

halo2noRT024, = ax.plot('no RT,\nz = 0.024', M_half_DM[0] / M_half_star[0], 's', color='#3185FC', markersize=11)
plt.plot('no RT,\nz = 0.024', M_half_x_wolf[0] / M_half_star[0], 's', color='#083D77', markersize=11)
plt.plot('no RT,\nz = 0.024', M_half_y_wolf[0] / M_half_star[0], 's', color='#DA4167', markersize=11)
plt.plot('no RT,\nz = 0.024', M_half_z_wolf[0] / M_half_star[0], 's', color='#F4D35E', markersize=11)

halo2RT024, = ax.plot('with RT,\nz = 0.024', M_half_DM[1] / M_half_star[1], '^', color='#3185FC', markersize=11)
plt.plot('with RT,\nz = 0.024', M_half_x_wolf[1] / M_half_star[1], '^', color='#083D77', markersize=11)
plt.plot('with RT,\nz = 0.024', M_half_y_wolf[1] / M_half_star[1], '^', color='#DA4167', markersize=11)
plt.plot('with RT,\nz = 0.024', M_half_z_wolf[1] / M_half_star[1], '^', color='#F4D35E', markersize=11)

halo2noRT130, = ax.plot('no RT,\nz = 0.130', M_half_DM[2] / M_half_star[2], 'o', color='#3185FC', markersize=11)
plt.plot('no RT,\nz = 0.130', M_half_x_wolf[2] / M_half_star[2], 'o', color='#083D77', markersize=11)
plt.plot('no RT,\nz = 0.130', M_half_y_wolf[2] / M_half_star[2], 'o', color='#DA4167', markersize=11)
plt.plot('no RT,\nz = 0.130', M_half_z_wolf[2] / M_half_star[2], 'o', color='#F4D35E', markersize=11)

halo2RT130, = ax.plot('with RT,\nz = 0.130', M_half_DM[3] / M_half_star[3], 'P', color='#3185FC', markersize=11)
plt.plot('with RT,\nz = 0.130', M_half_x_wolf[3] / M_half_star[3], 'P', color='#083D77', markersize=11)
plt.plot('with RT,\nz = 0.130', M_half_y_wolf[3] / M_half_star[3], 'P', color='#DA4167', markersize=11)
plt.plot('with RT,\nz = 0.130', M_half_z_wolf[3] / M_half_star[3], 'P', color='#F4D35E', markersize=11)

halo2noRT373, = ax.plot('no RT,\nz = 0.373', M_half_DM[4] / M_half_star[4], 'd', color='#3185FC', markersize=11)
plt.plot('no RT,\nz = 0.373', M_half_x_wolf[4] / M_half_star[4], 'd', color='#083D77', markersize=11)
plt.plot('no RT,\nz = 0.373', M_half_y_wolf[4] / M_half_star[4], 'd', color='#DA4167', markersize=11)
plt.plot('no RT,\nz = 0.373', M_half_z_wolf[4] / M_half_star[4], 'd', color='#F4D35E', markersize=11)

halo2RT373, = ax.plot('with RT,\nz = 0.373', M_half_DM[5] / M_half_star[5], 'h', color='#3185FC', markersize=11)
plt.plot('with RT,\nz = 0.373', M_half_x_wolf[5] / M_half_star[5], 'h', color='#083D77', markersize=11)
plt.plot('with RT,\nz = 0.373', M_half_y_wolf[5] / M_half_star[5], 'h', color='#DA4167', markersize=11)
plt.plot('with RT,\nz = 0.373', M_half_z_wolf[5] / M_half_star[5], 'h', color='#F4D35E', markersize=11)

#plt.xlabel('M$_{half,stars}$ (M$_\odot$)', fontsize=17)
plt.ylabel(r'$\frac{M_{half,DM}}{M_{half,stars}}$', fontsize=24) # using LATEX here!

plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
patch1 = mpatches.Patch(color='#3185FC', label='DM halo')
patch2 = mpatches.Patch(color='#083D77', label='Wolf, x')
patch3 = mpatches.Patch(color='#DA4167', label='Wolf, y')
patch4 = mpatches.Patch(color='#F4D35E', label='Wolf, z')

plt.legend(handles=[patch1, patch2, patch3, patch4], fontsize=14, loc='lower left')

plt.title('Ratio between M$_{half,DM}$ and M$_{half,stars}$\nwith respect to M$_{half,stars}$, dwarf 2', \
         fontsize=19)

plt.yscale('log')

plt.savefig('ratioplot2Wolf.png')
```
